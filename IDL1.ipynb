```python
class Trabajador:
    def __init__(self, nombre, resumen, estado, años_experiencia=0):
        self.nombre = nombre
        self.resumen = resumen
        self.estado = estado
        self.años_experiencia = años_experiencia
        self.jefe_inmediato = None

    def get_resumen(self):
        return f"Nombre: {self.nombre}, Resumen: {self.resumen}, Años de Experiencia: {self.años_experiencia}"

    def get_jefe_inmediato(self):
        if self.jefe_inmediato:
            return self.jefe_inmediato.nombre
        else:
            return "No tiene jefe inmediato"

    def get_estado(self):
        return f"Estado: {self.estado}"

    def set_jefe_inmediato(self, jefe):
        self.jefe_inmediato = jefe

class Gerente(Trabajador):
    def __init__(self, nombre, resumen, estado):
        super().__init__(nombre, resumen, estado)

class JefeArea(Trabajador):
    def __init__(self, nombre, resumen, estado, años_experiencia):
        super().__init__(nombre, resumen, estado, años_experiencia)
        self.personal_tecnico = []
        self.asistentes = []

    def add_personal_tecnico(self, tecnico):
        if len(self.personal_tecnico) < 5:
            self.personal_tecnico.append(tecnico)
            tecnico.set_jefe_inmediato(self)
        else:
            print("No se pueden añadir más técnicos a esta área")

    def add_asistente(self, asistente):
        if len(self.asistentes) < 2:
            self.asistentes.append(asistente)
            asistente.set_jefe_inmediato(self)
        else:
            print("No se pueden añadir más asistentes a esta área")

class PersonalTecnico(Trabajador):
    def __init__(self, nombre, resumen, estado, años_experiencia):
        super().__init__(nombre, resumen, estado, años_experiencia)

class Asistente(Trabajador):
    def __init__(self, nombre, resumen, estado):
        super().__init__(nombre, resumen, estado)

# Crear Gerente
gerente = Gerente("Gerente General", "Gerente de toda la empresa", "Activo")

# Crear Jefes de Área
jefe_marketing = JefeArea("Jefe de Marketing", "Encargado de Marketing", "Activo", 10)
jefe_sistemas = JefeArea("Jefe de Sistemas", "Encargado de Sistemas", "Activo", 8)
jefe_produccion = JefeArea("Jefe de Producción", "Encargado de Producción", "Activo", 12)
jefe_logistica = JefeArea("Jefe de Logística", "Encargado de Logística", "Activo", 7)

# Crear Personal Técnico
tecnico1 = PersonalTecnico("Técnico 1", "Técnico en Sistemas", "Activo", 5)
tecnico2 = PersonalTecnico("Técnico 2", "Técnico en Producción", "Activo", 3)

# Crear Asistentes
asistente1 = Asistente("Asistente 1", "Asistente en Logística", "Activo")
asistente2 = Asistente("Asistente 2", "Asistente en Marketing", "Activo")

# Añadir Personal a Jefes de Área
jefe_sistemas.add_personal_tecnico(tecnico1)
jefe_produccion.add_personal_tecnico(tecnico2)
jefe_logistica.add_asistente(asistente1)
jefe_marketing.add_asistente(asistente2)

# Crear el Array de Objetos
trabajadores = [gerente, jefe_marketing, jefe_sistemas, jefe_produccion, jefe_logistica, tecnico1, tecnico2, asistente1, asistente2]

# Función para listar en Consola
def listar_trabajadores(trabajadores):
    for trabajador in trabajadores:
        print(f"Nombre: {trabajador.nombre}")
        print(trabajador.get_resumen())
        print(f"Jefe Inmediato: {trabajador.get_jefe_inmediato()}")
        print(trabajador.get_estado())
        print("-" * 40)

# Listar los Trabajadores
listar_trabajadores(trabajadores)
```
